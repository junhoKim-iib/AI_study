```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris # iris 데이터 로드
from sklearn.tree import DecisionTreeClassifier,plot_tree

iris = load_iris() # iris 데이터 로드

# Decision Tree 모델 객체 생성
decision_tree = DecisionTreeClassifier(criterion="entropy" # entropy를 불순도 계산방법으로 적용. 
                                       #(criterion : 분할 품질을 측정하는 기능)
                                       ,random_state=0  # 난수 seed 설정
                                       ,max_depth=4  #트리의 최대 깊이
                                      )
# Decision Tree 모델 객체 학습
decision_tree = decision_tree.fit(iris.data #훈련과 테스트 특성
                                  , iris.target) #훈련과 테스트 타깃

# 출력
plt.figure()
plot_tree(decision_tree, filled=True)
plt.show()
```
