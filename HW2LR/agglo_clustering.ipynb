```python
from sklearn.datasets import make_blobs
from scipy.spatial.distance import pdist
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
import numpy as np
from matplotlib import pyplot as plt
```

```python
#  Generate isotropic Gaussian blobs for clustering.
X, y = make_blobs(n_samples=125, n_features=2, centers=3, cluster_std=2.5)
colormap = np.array(['r', 'g', 'b'])
plt.scatter(X[:,0], X[:,1], c=colormap[y])
plt.title('Genreated Data')
plt.show()
```

```python
'''
scipy.spatial.distance.pdist(X, metric='euclidean', *, out=None, **kwargs)

  Pairwise distances between observations in n-dimensional space.

parameters :

# X : array_like  / An m by n array of m original observations in an n-dimensional space.
# metric : str or function, optional  / The distance metric to use.

etc...

Returns:
# Y: ndarray  / Returns a condensed distance matrix Y.
'''
Xdist = pdist(X, metric='euclidean')  # calculate distance matrix

'''
scipy.cluster.hierarchy.linkage(y, method='single', metric='euclidean', optimal_ordering=False)

  Perform hierarchical/agglomerative clustering.

parameters:

# y: ndarray  / A condensed distance matrix.
# method: str, optional  / The linkage algorithm to use. See the Linkage Methods section below for full descriptions.
'''
Z = linkage(Xdist, method= 'ward')  # ward linkage


Zd = dendrogram(Z)
plt.title('Dendrogram')
plt.show()
```
